|proprietà|FIAT|OPEL|

Per ottenere lo scarto quadrato medio si usa la formula:

<p>\(V(Y)=\frac{1}{n}\sum_{i=1}^{n}(y_i-E(Y))^2  \)</p>
<p>Sostuidendo con i valori:</p>
<p>Autovetture FIAT:</p>
<p>\(V(Y)=\frac{1}{40}\cdot(  (9\cdot(0-1,5)^2)  + (13\cdot(1-1,5)^2)  +(10\cdot(2-1,5)^2)  +(5\cdot(3-1,5)^2)  +(3\cdot(4-1,5)^2) )=1,4     \)</p>
<p>Per il coeficente di variazione:</p>
<p>\( σ_A = \sqrt{1,4}≈1,183 \)</p>
<p>\(  CV=\frac{σ}{\overline{Y}}\cdot100 = \frac{1,183}{1,5}\cdot100=78,9\% \)</p>
<p>Autovetture OPEL:</p>
<p>\(V(v)=\frac{1}{60}\cdot(33\cdot(0-0,583)^2 + 20\cdot(1-0,583)^2 + 6\cdot(2-0,583)^2+ 1\cdot(3-0,583)^2+ 0\cdot(4-0,583)^2)≈0,543   \)</p>
<p>Per il coeficente di variazione:</p>
<p> \( σ_A = \sqrt{0,543}≈0.737 \) </p>
<p>\(  CV=\frac{σ}{\overline{Y}}\cdot100 = \frac{0.737}{0,583}\cdot100=126,4\% \)</p>

Dall’analisi dei dati emerge che le autovetture OPEL presentano un numero medio di guasti inferiore (0,583 rispetto a 1,5 per le autovetture FIAT), risultando quindi mediamente più affidabili. Tuttavia, il loro coefficiente di variazione (126,4%) è maggiore rispetto a quello delle autovetture FIAT (78,9%), il che indica che la distribuzione dei guasti tra i diversi veicoli OPEL è più disomogenea: alcune vetture non presentano alcun guasto, mentre altre mostrano più problemi.
In sintesi, le autovetture OPEL sono mediamente più affidabili, ma mostrano una maggiore variabilità individuale nei risultati rispetto alle FIAT.
